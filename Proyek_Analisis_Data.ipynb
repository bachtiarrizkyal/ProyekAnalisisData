### Pertanyaan 1: Produk apa yang paling laris dan kurang diminati?

```python
sum_order_items_df = all_data.groupby("product_category_name_english")["product_id"].count().reset_index()
sum_order_items_df = sum_order_items_df.rename(columns={"product_id": "products"})
sum_order_items_df = sum_order_items_df.sort_values(by="products", ascending=False)

sum_order_items_df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#3366CC", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="products", y="product_category_name_english", data=sum_order_items_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Produk paling laris", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

sns.barplot(x="products", y="product_category_name_english", data=sum_order_items_df.sort_values(by="products", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Produk paling kurang diminati", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle("Produk Paling Laris dan Kurang Diminati", fontsize=20)
plt.show()
```

Dapat dilihat pada grafik diatas, **produk yang paling laris adalah bed_bath_table**, sedangkan **produk yang paling kurang diminati adalah security_and_services**.

### Pertanyaan 2: Bagaimana penjualan di platform E-Commerce?

```python
# Mengelompokkan data berdasarkan bulan dan menghitung jumlah order unik setiap bulannya
monthly_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})

# Mengubah indeks tanggal menjadi nama bulan dalam format teks (e.g., 'January', 'February', dst.)
monthly_df.index = monthly_df.index.strftime('%B')

# Mereset indeks DataFrame dan mengubah nama kolom order_id menjadi order_count
monthly_df = monthly_df.reset_index()
monthly_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)

# Menampilkan DataFrame
monthly_df
```

```python
# Mengurutkan DataFrame berdasarkan kolom order_count dan menjaga baris dengan nilai terakhir untuk setiap order_approved_at
monthly_df = monthly_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
monthly_df
```

```python
# Dictionary untuk memetakan nama bulan ke nilai numerik
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

# Menambahkan kolom baru 'month_numeric' yang merupakan nilai numerik dari nama bulan
monthly_df["month_numeric"] = monthly_df["order_approved_at"].map(month_mapping)

# Mengurutkan DataFrame berdasarkan nilai 'month_numeric'
monthly_df = monthly_df.sort_values("month_numeric")

# Menghapus kolom 'month_numeric' karena sudah tidak diperlukan lagi
monthly_df = monthly_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(
    monthly_df["order_approved_at"],
    monthly_df["order_count"],
    marker='o',
    linewidth=2,
    color="#3366CC",
    linestyle='-',
    mec='black',
    mew=1
)
plt.title("Penjualan Pada Tahun 2023", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

Pada grafik diatas dapat dilihat bahwa penjualan **paling sedikit terdapat dalam bulan September**, dan **meningkat secara signifikan pada bulan November**. Serta **penjualan terbanyak terdapat pada bulan November**.

```python
# Menghitung jumlah masing-masing nilai dalam kolom 'review_score', kemudian mengurutkannya secara menurun
review_scores = all_data['review_score'].value_counts().sort_values(ascending=False)

# Mendapatkan nilai paling umum dari kolom 'review_score'
most_common_scores = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5))
sns.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            palette=["#068DA9" if score == most_common_scores else "#D3D3D3" for score in review_scores.index]
            )

plt.title("Tingkat Kepuasan Customer", fontsize=20)
plt.xticks(fontsize=12)
plt.show()
```

Pada grafik diatas dapat dilihat bahwa **customer sangat puas**. Hal ini dapat dilihat bahwa customer paling banyak memberikan nilai 5/5.
